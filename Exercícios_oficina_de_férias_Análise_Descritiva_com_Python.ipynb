Descobrir quantos segundos tem uma semana.

```python
transforma_semanas_em_segundos(1)
```



---



```python
def transforma_semanas_em_segundos (semanas_solicitadas):
  dias_por_semana = 7
  horas_por_dia = 24
  minutos_por_hora = 60
  segundos_por_minuto = 60
  return semanas_solicitadas * dias_por_semana * horas_por_dia * minutos_por_hora * segundos_por_minuto
```

# Exercício 02

Descobrir se o número informado é par ou impar.

```python
def numero_par_ou_impar (numero):
  if (numero % 2 == 0):
    print(f'O número {numero} é par.')
    return
  print(f'O número {numero} é ímpar.')
```

# Exercício 03

Leitura de base CSV.

```python
import pandas as pd
df = pd.read_csv('tips.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Estatísticas descritivas
descriptive_stats = df.describe()
print("Estatísticas Descritivas:\n", descriptive_stats, "\n")

# Gráfico de dispersão: total_bill vs tip
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["total_bill"], y=df["tip"], hue=df["sex"], style=df["smoker"], s=100)
plt.xlabel("Total da Conta")
plt.ylabel("Gorjeta")
plt.title("Relação entre Total da Conta e Gorjeta")
plt.legend(title="Sexo")
plt.show()
print("\n")

# Boxplot para análise de gorjetas por sexo
plt.figure(figsize=(6, 4))
sns.boxplot(x=df["sex"], y=df["tip"], legend=False, hue=df["sex"], palette="pastel")
plt.xlabel("Sexo")
plt.ylabel("Gorjeta")
plt.title("Distribuição de Gorjetas por Sexo")
plt.show()
print("\n")

# Média da gorjeta por sexo
mean_tips_by_sex = df.groupby("sex")["tip"].mean()
print("Média das gorjetas por sexo:\n", mean_tips_by_sex)
print("\n")

# Gráfico dinâmico de total de vendas por dia
import plotly.express as px

daily_sales = df.groupby("sex")["total_bill"].sum().reset_index()
fig = px.bar(daily_sales, x="sex", y="total_bill", title="Total de Vendas por Dia", labels={"total_bill": "Total de Vendas", "sex": "Dia"})
fig.show()
```

# Exercício 08

Fundamentos da análise de dados, utilizando base de dados das [Vendas do Tesouro Direto](https://drive.google.com/file/d/1cfQ6MQaWfuiqropirimXJ7zaq5IEbC6X/view?usp=sharing).
https://www.tesourotransparente.gov.br/ckan/dataset/vendas-do-tesouro-direto/resource/e5f90e3a-8f8d-4895-9c56-4bb2f7877920

Dicionário de dados da base.

| Coluna                  | Descrição em Português                          | Tipo em Python |
|-------------------------|------------------------------------------------|----------------|
| `Tipo Titulo`           | Tipo do título de tesouro                      | `str`          |
| `Vencimento do Titulo`  | Data de vencimento do título                   | `str` (ou `datetime` se convertido) |
| `Data Venda`            | Data da venda do título                        | `str` (ou `datetime` se convertido) |
| `PU`                    | Preço Unitário do título                       | `float`        |
| `Quantidade`            | Quantidade de títulos negociados               | `float`        |
| `Valor`                 | Valor total da operação                        | `float`        |

```python
!pip install sweetviz
import pandas as pd
import sweetviz as sv
```

```python
# Leitura dos dados
df = pd.read_csv('VendasTesouroDireto.csv', sep=';')
print(">>> Leitura dos dados realizada com sucesso!!\n\n")

# Limpeza dos dados nulos
df = df.dropna()
print(">>> Limpezas de dados nulos realizada com sucesso!!\n\n")

# Transforma float inválido em valor válido.
def invalidStringFloatToFloat(invalid_string):
  return invalid_string.replace(to_replace=r'\,', value='.', regex=True).astype('float')

# Tratamento dos dados quantitativos
df['PU'] = invalidStringFloatToFloat(df['PU'])
df['Quantidade'] = invalidStringFloatToFloat(df['Quantidade'])
df['Valor'] = invalidStringFloatToFloat(df['Valor'])
print(">>> Tratamento dos dados quantitativos realizado com sucesso!!\n\n")

# # Tratamento dos dados de tempo
df['Vencimento do Titulo'] = df['Vencimento do Titulo'].as
df['Data Venda'] = df['Data Venda']
# print(">>> Tratamento dos dados quantitativos realizado com sucesso!!\n\n")

# Apresentação dos primeiros dados
print(">>> Preview dos dados:\n")
print(df.head(), '\n\n')

# Apresentação das informações gerais dos dados
print(">>> Informações dos dados:\n")
print(df.info(), '\n\n')

# Estatísticas descritivas dos dados
print(">>> Estatísticas descritivas:\n")
estatisticas_descritivas = df.describe(include='all')
print(estatisticas_descritivas, '\n\n')

# Gerando estatísticas descritivas em formato CSV
estatisticas_descritivas.to_csv('estatisticas_descritivas_VendasTesouroDireto.csv')
print(">>> CSV das estatísticas descritivas realizada com sucesso!!\n\n")
```

```python
# Gerando relatório do sweetviz
print(">>> Análise dos dados com sweetviz\n")
report = sv.analyze(df)
report.show_html("sweetviz_report_VendasTesouroDireto.html")
print(">>> HTML do relatório sweetviz gerado com sucesso!!\n\n")
```
